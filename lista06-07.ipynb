## 1-) Defina a função soma_nat que recebe como argumento um número natural n e devolve a soma de todos os números naturais até n.

## LAMBDA

```python
#funcao
def soma_nat(n):
    soma = 0
    while(n > 0):
        soma = soma + n
        n = n - int(1)
    return soma


#lambda
soma_nat_lambda = lambda end: ((end)/2) * (end+1)


#Maybe monad
def soma_nat_maybe(end):
    return ((end)/2) * (end+1)

class Just:
    
    def __init__(self,end):
        
        self.end = end
    def bind(self,fnc):
        try:
            return Just(fnc(self.end))
        except:
            return Nothing()
        
    def __repr__(self):
        
        return str(self.end)
```

## 2-) Defina a função prod_lista que recebe como argumento uma lista de inteiros e devolve o produto dos seus elementos.

## LAMBDA

```python
#funcao
def prod_lista(lista):
    soma = 1
    for i in lista:
        soma = soma * i
    return soma

#lambda
import operator
from functools import reduce
prod_lista_lambda = lambda x: reduce(operator.mul, x, 1)


#Maybe monad
def soma_list_maybe(x):
    return reduce(operator.mul, x, 1)

class Just:
    
    def __init__(self,x):
        
        self.x = x
    def bind(self,fnc):
        try:
            return Just(fnc(self.x))
        except:
            return Nothing()
        
    def __repr__(self):
        
        return str(self.x)
```

## 3-) Defina a função contem_parQ que recebe como argumento uma lista de números inteiros w e devolve True se w contém um número par e False em caso contrário.

## LAMBDA

```python
#funcao
def parQ(lista):
    if len(lista)%2==0:
        return True
    else:
        return False
    
#lambda
parQ_lambda = lambda lista: True if (len(lista) % 2 == 0) else False


#Maybe monad
def parQ_maybe(lista):
    return True if (len(lista) % 2 == 0) else False

class Just:
    
    def __init__(self,lista):
        
        self.lista = lista
    def bind(self,fnc):
        try:
            return Just(fnc(self.lista))
        except:
            return Nothing()
        
    def __repr__(self):
        
        return str(self.lista)
```

## 4-) Defina a função todos_imparesQ que recebe como argumento uma lista de números inteiros w e devolve True se w contém apenas números ímpares e False em caso contrário.

## LAMBDA

```python
#Funcao
def todos_imparesQ(lista):
    retorno = False
    for i in lista:
        
        if i%2 > 0:
            retorno = True
        else:
            retorno = False
            break
    return retorno


#lambda
todos_imparesQ_lambda = lambda lista: True if len([x for x in lista if x % 2 != 0])==len(lista) else False

#Maybe monad
def todos_imparesQ_maybe(lista):
    return True if len([x for x in lista if x % 2 != 0])==len(lista) else False

class Just:
    
    def __init__(self,lista):
        
        self.lista = lista
    def bind(self,fnc):
        try:
            return Just(fnc(self.lista))
        except:
            return Nothing()
        
    def __repr__(self):
        
        return str(self.lista)
```

## 5-) Defina a função inverteLista que recebe como argumento uma lista w e devolve a mesma lista mas invertida.

## LAMBDA

```python
#funcao
def inverteLista(w):
    w = w[::-1]
    return w

#lambda
inverteLista_lambda = lambda w: w[::-1]


#Maybe monad
def inverteLista_maybe(w):
    w = w[::-1]
    return w

class Just:
    
    def __init__(self,w):
        
        self.w = w
    def bind(self,fnc):
        try:
            return Just(fnc(self.w))
        except:
            return Nothing()
        
    def __repr__(self):
        
        return str(self.w)
class Nothing:
    
    def bind(self,fnc):
        
        return Nothing()
    
    def __repr__(self):
        
        return 'Nothing'
```
